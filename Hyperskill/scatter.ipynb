```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from vega_datasets import data
```

## Scatter Plot

Scatter plot is used to show the relationship between two variables/features. In this notebook I will try to check the relationship between temperature and wind.

```python
# loading the dataset
weather = data.seattle_weather()
weather.head()
```

```python
# feature engineering
weather["temp"] = np.round((weather.temp_max + weather.temp_min) / 2, 1)
weather.head()
```

Drawing the scatter plot

```python
plt.figure(figsize=(12, 6))
plt.scatter(weather.temp, weather.wind)
```

The default marker type is circle, but we can change it.

```python
plt.figure(figsize=(12, 6))
plt.scatter(weather.temp, weather.wind, marker="x")
```

We can also assign colors based on either of the features

```python
plt.figure(figsize=(12, 6))
plt.scatter(weather.temp, weather.wind, c=weather.temp)
```

We can see the meaning of the colors from the graph, as the temperature increases the color will get brighter. But to make it more meaningful, we can add a color bar.

```python
plt.figure(figsize=(12, 6))
plt.scatter(weather.temp, weather.wind, c=weather.temp)
plt.colorbar()
```

As I mentioned earlier, we can assign the color values to the other variable like:

```python
plt.figure(figsize=(12, 6))
plt.scatter(weather.temp, weather.wind, c=weather.wind)
plt.colorbar()
```

we can also arrange the size of the markers

```python
plt.figure(figsize=(12, 6))
plt.scatter(weather.temp, weather.wind, c=weather.wind, s=weather.wind * 2)
plt.colorbar()
```

As we can see above, as the value for wind increase the size of the markers will also get increase.

We can also categorize points using other categorical feature. In this case let's use `weather` feature from our dataset but let's first map the feature into colors, without mapping we don't get the intended results.

```python
weather.weather.unique()
```

```python
colors = weather.weather.map({"drizzle": "blue", "rain": "green", "snow": "red", "fog": "yellow", "sun": "violet"})
```

```python
plt.figure(figsize=(12, 6))
plt.scatter(weather.temp, weather.wind, c=colors, alpha=0.5)
plt.colorbar()
```

We can also use, `plt.plot()` function to plot scatter plots

```python
plt.plot(weather.temp, weather.wind, "o")
```

The `plt.plot()` function doesn't have an adequate methods to customize the graph
